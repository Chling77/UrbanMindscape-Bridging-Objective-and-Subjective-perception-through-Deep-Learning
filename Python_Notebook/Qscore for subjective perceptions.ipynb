```python
# 使用pandas的read_csv函数，指定分隔符为制表符
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Data/place-pulse-2.0/qscores.tsv', sep='\t')

df
```

```python
study = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Data/place-pulse-2.0/studies.tsv', sep='\t')
study
```

```python
# 首先，创建一个映射字典，键是study表的id，值是study_question。
study_question_mapping = pd.Series(study.study_question.values,index=study._id).to_dict()

# 然后，使用这个映射来创建一个新列。
df['study_question'] = df['study_id'].map(study_question_mapping)

df
```

```python
df.to_csv('/content/drive/My Drive/Colab Notebooks/Data/new_qscores.csv', index=False)
```
